```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything with other feature
```

```python
def correlation(dataset,threshold):
    col_corr =set() # set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr    
```

```python
corr_features=correlation(db_data,0.8)
len(set(corr_features))               
```

```python
 # if we take threshold 0.8,there is no highly correlated variable.hence we dont drop the columns
```

# TRAIN_TEST_SPLIT MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
models=[LogisticRegression(max_iter=1000),KNeighborsClassifier(n_neighbors=5),DecisionTreeClassifier(),RandomForestClassifier(),SVC(kernel='linear')]
```

# comparing performance of models using a function

```python
def compare_models_train_test():
    for algorithm in models:
        algorithm.fit(X_train,y_train)
        prediction=algorithm.predict(X_test)
        accuracy=accuracy_score(y_test,prediction)
        print("accuracy score of the",algorithm,"=",accuracy)
        
```

```python
compare_models_train_test()
```

```python
# observations
here in the train_test_split model ,
highest accuracy score identified for logistic regression algorithm with 73%
min accuracy score identified for decision tree classifier with 68.8%

there is not much difference in accuracies when we compared between  cross validation and train_test_split models
so, we can consider "logistic regression" algorithm to build a model for the given data set
```

# comparing for both train and test data set

```python
def compare_models_train_test():
    for algorithm in models:
        algorithm.fit(X_train,y_train)
        prediction_train=algorithm.predict(X_train)
        prediction_test=algorithm.predict(X_test)
        accuracy_train=accuracy_score(y_train,prediction_train)
        accuracy_test=accuracy_score(y_test,prediction_test)
        print("accuracy score of the train set",algorithm,"=",accuracy_train)
        print("accuracy score of the test set",algorithm,"=",accuracy_test)
        print(" --------------------------------------------------------------------------")
        
```

```python
compare_models_train_test()
```

# precision score
precision= TP/(TP+FP)
it means how well the model perform regarding prediction of the "positive class"
i.e, out of total predicted positives,how many were correctly predicted as positive.

```python
print("precision score on train data:\n", precision_score(y_train,my_logreg_preds_train))
```

```python
print("precision score on test data:\n", precision_score(y_test,my_logreg_preds_test))
```

# recall score
recall=TP/(TP+FN)
it means, out of total actual positives,how many are true.
how many instances were correctly identified as a people who have a disease

```python
print("recall score on train data:\n", recall_score(y_train,my_logreg_preds_train))
```

```python
print("recall score on test data:\n", recall_score(y_test,my_logreg_preds_test))
```
